```python
#브랜빕샘플명 추출 및 중복 샘플 제거 
def get_branvip_sample_name_list_and_remove_overlaped_list(
    src_path = 'sample_total_41893',
    compare_path = 'train',
     output_file = '브랜빕샘플리스트.txt',
      output_file2 = '브랜빕샘플중복리스트_train.txt'):

    import os

    compare_list = os.listdir(compare_path)
    number_list = [] 
    overlapped_list = [] 


    sample_list = os.listdir(src_path)
    for sample in sample_list:
        sample_temp = sample.split('_')
        #print(sample_temp[0],len(sample_temp[0]))
        
        #010102_1003329&기타&단순.jpg 경우
        if len(sample_temp[0]) == 6:
            
            #010102_1003329&기타&단순.jpg 경우  
            if '&' in sample:
                sample_temp2 = sample.split('&')
                #print(sample_temp2[0].split('_')[1])
                number_list.append(sample_temp2[0].split('_')[1])
            
            #010103_4020190118710_전통_친근_초록_초록 경우
            if '&' not in sample:
                #print(sample_temp[1])
                number_list.append(sample_temp[1])
        
        if len(sample_temp[0]) != 6:
            #print(sample_temp[0],33333)
            number_list.append(sample_temp[0])
            

    print('브랜빕샘플 = ',len(number_list))          
    # Write each file path to the output file
    # with open(output_file, "w") as f:
    #     for number in number_list:
    #         f.write(number + "\n")   
            
            
    #중복리스트 추출
    for compare_sample in compare_list:
        compare_temp = compare_sample.split('_')[1].split('.txt')[0]
        #print(compare_temp)
        if compare_temp in number_list:
            #print('중복 = ', compare_temp)
            overlapped_list.append(compare_temp)
            #중복삭제 
            os.remove(os.path.join(compare_path,compare_sample))
            
    print('브랜빕샘플 중복 = ',len(overlapped_list))          
    # Write each file path to the output file
    # with open(output_file2, "w") as f:
    #     for number in overlapped_list:
    #         f.write(number + "\n") 

get_branvip_sample_name_list_and_remove_overlaped_list() 
```
