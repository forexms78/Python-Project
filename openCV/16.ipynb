```python
#욜로이미지를 txt 박싱참조해서 crop 이미지로 컨버트후 save
def yolo_convert_image_to_cropped_image_and_save(
    src_path = 'train',
    dst_path = 'train_cropped'):

    import cv2
    import os
    import numpy as np

    # Function to extract bounding box info from a YOLO-formatted annotation file
    def read_yolo_annotation(txt_path):
        with open(txt_path, 'r') as f:
            line = f.readline().strip().split()
            if len(line) < 5:
                return None  # No bounding box info
            class_id, x_center, y_center, width, height = map(float, line)
            return class_id, x_center, y_center, width, height

    # File paths
    #txt_path = "010102_1013442.txt"
    #image_path = "010102_1013442.jpg"
    
    error_count = 0 

    txt_path_list = os.listdir(src_path)
    for txt in txt_path_list:
        if '.txt' in txt:
            txt_path = os.path.join(src_path,txt)
            #print(txt_path) 
            #break 
            
            # Read the annotation and image
            annotation = read_yolo_annotation(txt_path)
            if annotation is not None:
                class_id, x_center, y_center, width, height = annotation
                image_path = os.path.join(src_path,txt.replace('.txt','.jpg'))
                #print(image_path)
                cropped_image_path = os.path.join(dst_path,os.path.basename(image_path))
                #print(cropped_image_path)
                
            
                image = cv2.imread(image_path)

                # Convert normalized coordinates to actual pixel values
                img_height, img_width, _ = image.shape
                x_center, y_center, width, height = x_center * img_width, y_center * img_height, width * img_width, height * img_height

                # if width == 0 and height == 0:
                #     # Create a small dummy image and save
                #     dummy_image = np.zeros((10, 10, 3), dtype=np.uint8)
                #     cv2.imwrite(cropped_image_path, dummy_image)
                
                # else:
                
                
                
                # Calculate coordinates for cropping
                x1, y1 = int(x_center - width // 2), int(y_center - height // 2)
                x2, y2 = int(x_center + width // 2), int(y_center + height // 2)

                # Crop and save the image
                cropped_image = image[y1:y2, x1:x2]
                #print(cropped_image_path)
                try:
                    cv2.imwrite(cropped_image_path, cropped_image)
                except:
                    error_count += 1 
                    print(error_count)
                    

yolo_convert_image_to_cropped_image_and_save()
```
