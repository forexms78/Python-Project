```python
#카테고리 a foler -> b folder 10% 이동 
def yolo_move_10_percent(src_folder = "train",
                        dst_folder = "valid",
                        key_compare = '010102',
                        move_with_jpg = True):

    import os
    import random
    import shutil

    # Set the source and destination folder paths
    
 

    # Create the destination folder if it doesn't exist
    if not os.path.exists(dst_folder):
        os.mkdir(dst_folder)

    # Create a dictionary to hold filenames grouped by their first 6 characters
    file_groups = {}
    for filename in os.listdir(src_folder):
        if filename.endswith('.txt'):
            key = filename[:6]
        
            if key == key_compare:
                
                if key not in file_groups:
                    file_groups[key] = []
                file_groups[key].append(filename)
                #print(file_groups[key])
                



    # For each group, move 10% of the files to the destination folder
    for key, filenames in file_groups.items():
        num_to_move = max(1, int(len(filenames) * 0.1))  # At least 1 file will be moved
        files_to_move = random.sample(filenames, num_to_move)
        
        for filename in files_to_move:
            #print(filename)
            if move_with_jpg:
                src_path = os.path.join(src_folder, filename)
                dst_path = os.path.join(dst_folder, filename)
                
                shutil.move(src_path, dst_path)
                
                src_path_jpg = os.path.join(src_folder, filename.replace('.txt','.jpg'))
                dst_path_jpg = os.path.join(dst_folder, filename.replace('.txt','.jpg'))
                
                shutil.move(src_path_jpg, dst_path_jpg)
            else:
                src_path = os.path.join(src_folder, filename)
                dst_path = os.path.join(dst_folder, filename)
                
                shutil.move(src_path, dst_path)
            #break 
                 
            #print(f"Moved {filename} from {src_folder} to {dst_folder}")     

yolo_move_10_percent() 
```
